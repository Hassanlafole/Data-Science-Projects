```python
top_df = epl_df
top_df['Goals_Difference'] = top_df['Goals'] - top_df['xG']
top_df['Assists_Difference'] = top_df['Assists'] - top_df['xA']

# Find players who consistently outperformed xG and xA
consistent_outperformers = epl_df[
    (top_df['Goals_Difference'] > 0) & (top_df['Assists_Difference'] > 0)
].sort_values(by=['Goals_Difference', 'Assists_Difference'], ascending=False)

# Display the top players who consistently outperformed xG and xA
print(consistent_outperformers[['Name', 'Club', 'Goals_Difference', 'Assists_Difference']].head())
```

```python
club_stats = epl_df.groupby('Club').agg({'xG': 'mean', 'Goals': 'mean'}).reset_index()

club_stats['xG_Goals_Difference'] = club_stats['Goals'] - club_stats['xG']

threshold = 0.1

clubs_high_xg_low_goals = club_stats[club_stats['xG_Goals_Difference'] > threshold]

plt.figure(figsize=(12, 6))
plt.barh(clubs_high_xg_low_goals['Club'], clubs_high_xg_low_goals['xG_Goals_Difference'])
plt.xlabel('Difference between xG and Goals')
plt.ylabel('Club')
plt.title('Clubs with High xG but Low Goals')
plt.gca().invert_yaxis() 
plt.show()
```

```python
features=["Matches","Mins","Goals"]
x=epl_df[features]
y=epl_df["Starts"]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3)
```

```python
epl_df=LinearRegression()
epl_df.fit(xtrain,ytrain)
```

```python
print("Accuracy: ",epl_df.score(xtest,ytest))
```
