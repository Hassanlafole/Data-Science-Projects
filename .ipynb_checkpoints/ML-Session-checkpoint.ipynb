## Importing Lbraries

```python
import pandas as pd
import numpy as np
import plotly.express as pe
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# hadii scikit-learn uusan kuugu jirin faldan sidaan hoos ku xusan isticmaal
# !pip install scikit-learn  
```

### Reading data and ensuring dataset variables

```python
df=pd.read_csv("advertising.csv")
```

```python
df
```

## Trendline analysis for Sales as target and Radio as predictor

```python
fig=pe.scatter(data_frame=df,x="Sales",y="Radio",size="Radio",trendline="ols")
fig.show()
```

## Get x and y variables 

```python
features=["TV","Radio","Newspaper"]
x=df[features]
y=df["Sales"]
```

## sptilling data into trainset and testset

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3)
```

```python
ml=LinearRegression()
ml.fit(xtrain,ytrain)
```

## Model Evaulation

```python
print("Accuracy: ",ml.score(xtest,ytest))
```

```python
adv=[[47,89,106]]
print(ml.predict(adv))
```
