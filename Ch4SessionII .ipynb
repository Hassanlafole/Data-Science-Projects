## Reading important Libraries & External Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df=pd.read_csv("movies.csv")
```

```python
df.columns
```

## Visualizing data using Histigram  in seaborn

```python
sn.histplot(df["budget"],bins=10,kde=True)
plt.title("Budget Distribution")
plt.show()
```

## Film rating analysis using scatter chart in seaborn

```python
sn.scatterplot(data=df,x="imdb_rating",y="revenue")
plt.title("Film rating by revenue")
plt.show()
```

## Film rating analysis using line chart in seaborn

```python
sn.lineplot(data=df,x="imdb_rating",y="revenue",hue="language",palette="viridis")
plt.title("Film rating by revenue")
plt.show()
```

## Film writen languages

```python
sn.countplot(data=df,x="language")
plt.title("Film writen languages")
plt.show()
```

## hands-n

## Rating Distribution by Industry Box plots can represent the spread of ratings across different industries.


```python
plt.figure(figsize=(12, 8))
sn.boxplot(x='industry', y='imdb_rating', data=df)
plt.title('Rating Distribution by Industry')
plt.show()
```

## Distribution of industry Categories using pie chart.

```python
industry_counts = df['industry'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(industry_counts, labels=industry_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Industry Categories')
plt.show()
```
