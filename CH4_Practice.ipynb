```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Airline Dataset.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
data=df["Gender"].value_counts()
label=df["Gender"].unique()

fig,ex = plt.subplots()
ex.pie(data,labels=label,autopct="%1.1f%%", startangle=45)
plt.show()
```

```python
nationality=df["Nationality"].value_counts()
nationality[:10].plot(kind="bar", color="blue")
plt.title("Top 10 most nationalities visited")
plt.xlabel("nationality")
plt.ylabel("Values")
plt.xticks(rotation=60)
plt.show()
```

```python
Gender_distribution=df.groupby(["Continents","Gender"]).size().unstack()
Gender_distribution.plot(kind="bar",stacked=True,color=["green","blue"])
plt.title("Gender Distribution by continents")
plt.xlabel("Continents")
plt.ylabel("Counts")
plt.show()
```

```python
FlightStatus=df["Flight Status"].value_counts()
FlightStatus.plot(kind="bar",color="yellow")
plt.title("Flight Status")
plt.xlabel("Status")
plt.ylabel("Counts")
plt.xticks(rotation=60)
plt.show()
```

```python
df["Departure Date"]=pd.to_datetime(df["Departure Date"])
```

```python
DOW=df["Departure Date"].dt.dayofweek.value_counts().sort_index()
DOW.plot(kind="line")
plt.title("Departure Days")
plt.xlabel("Days")
plt.ylabel("Counts")
plt.xticks(range(7),["Mon","Tue","Wed","Thu","Fri","Sat","Sun",])
plt.show()
```
